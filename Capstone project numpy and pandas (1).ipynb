# Task 2
The cost column in the dataframe “Project” has some missing values. Your task is to
compute these missing values. Replace the missing values by running average. You
should use the “For” loop for this task.


```python
import numpy as np   # importing numpy lib 

project = pd.read_csv('Project.csv')  # loading dataframe 
mean_cost = project['Cost'].mean()  # Calculate mean cost for non-missing values
for index, row in project.iterrows():  #filling the missing value
    if pd.isna(row['Cost']):
        project.loc[index, 'Cost'] = mean_cost
project.to_csv('Project.csv', index=False) #saving to csv 
project
```

# Task 3
Split the name column in the Employee dataframe into two new columns “First Name”,
and “Last Name” and remove the older “name” column.

```python
import pandas as pd  # importing pandas lib 

Employee = pd.read_csv('Employee.csv')  #loading df
Employee['First_Name'] = Employee['Name'].apply(lambda x: x.split()[0])  #creating firstname column
Employee['Last_Name'] = Employee['Name'].apply(lambda x: x.split()[1] if len(x.split()) > 1 else '') # creating lastname column
Employee.drop('Name', axis=1, inplace=True)  #droping the column name
Employee.to_csv('Employee.csv', index=False) #update the csv
Employee  # print the csv
```

# Task 4
Join all three dataframes in one single dataframe. Name it “Final”

```python
# Joining all data frames to one single frame
Project = pd.read_csv('Project.csv')  #loading df
Employee = pd.read_csv('Employee.csv')  #loading df
Seniority_Level = pd.read_csv('Seniority_Level.csv') #loading df

Final = Project.merge(Employee, on='ID', how='left')   # merging project df to employee on id 
Final = Final.merge(Seniority_Level, on='ID', how='left') # merging merged df to seniority df on id 
Final.to_csv('Final.csv', index=False) # save to csv
Final #print
```

# Task 5
Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project
cost only to employees who have finished the projects.

```python
Final= pd.read_csv('Final.csv') #loading df
Final['Bonus'] = Final['Cost'] * 0.05 * (Final['Status'] == 'Finished').astype(int)   # Add Bonus column
Final.to_csv('Final.csv', index=False)  # Save updated DataFrame
Final  #print the DataFrame
```

# Task 6
Demote the designation level by 1, whose projects have status “fail”. Delete the
employees record whose designation level is above 4.

```python
Final= pd.read_csv('Final.csv')  #loading df
failed_projects = Final.groupby('ID')['Status'].apply(lambda x: (x == 'Failed').sum()).reset_index(name='Failed_Projects')
Final = Final.merge(failed_projects, on='ID', how='left')
def update_designation(row):   #updating designation
    if row['Failed_Projects'] > 0:
        return max(row['Designation Level'] + 1, 5)  # Cap at 5
    return row['Designation Level']
Final['New_Designation_Level'] = Final.apply(update_designation, axis=1)
Final = Final[Final['New_Designation_Level'] <= 4]  # Filter out ineligible employees
Final.to_csv('Final.csv', index=False)  # save to csv
Final  #print
```
